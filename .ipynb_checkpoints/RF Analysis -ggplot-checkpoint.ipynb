```r
sigs <- read.csv('signals.csv', header=T, stringsAsFactors=F)
names(sigs) <- c('date', 'signal')
sigs$date <- as.POSIXct(sigs$date)
dewpts <- read.csv('dewpoints.csv')
names(dewpts) <- c('date', 'dewpoint')
dewpts$date <- as.POSIXct(dewpts$date)
```

```r
plt <- ggplot(sigs, aes(date, signal), group=1) 
plt <- plt + geom_line() + labs(x="", y="Signal (dBm)")
plt <- plt + labs(title='Hourly Signal Strength') +labs(subtitle="October 2, 1013 - June 24, 2016")
plt
```

```r
plt <- ggplot()
plt <- plt + geom_line(data=dewpts, aes(x=date, y=dewpoint)) + labs(x="", y="")
plt <- plt + labs(title='Hourly Dewpoint') + labs(subtitle="October 2, 1013 - June 24, 2016")
plt
```
