```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
import datetime
times = []
sigs = []
lines = open('signals.csv').readlines()
for line in lines[1:]:
    d,s = line.strip().split(',')
    try:
        date = datetime.datetime.strptime(d, "%Y-%m-%d %H:%M:%S")
    except:
        date = datetime.datetime.strptime(d, "%Y-%m-%d %H:%M:%S.%f")
    if date < datetime.datetime(2013,10,2):
        continue
    times.append(date)
    sigs.append([float(s)])
```

```python
plt.figure(figsize=[16,10])
plt.plot(times, sigs)
```

```python
dates = []
dewpoints = []
dewps = open('dewpoints.csv').readlines()
for dewp in dewps:
    d,s = dewp.strip().split(',')
    if float(s) < -50:
        continue
    try:
        date = datetime.datetime.strptime(d, "%Y-%m-%d %H:%M:%S")
    except:
        date = datetime.datetime.strptime(d, "%Y-%m-%d %H:%M:%S.%f")
    if date < datetime.datetime(2013,10,2):
        continue
    dates.append(date)
    dewpoints.append([float(s)])
```

```python
plt.figure(figsize=[16,10])
plt.plot(dates, dewpoints)
```
